```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

### Activate Function : Sigmoid

$\sigma(x)=\frac{1}{1+e^{-x}}$

$\sigma(x)'=f(x)(1-f(x))$

```python
def sigmoid(x):
    return(1/(1+np.exp(-x)))

def dsigmoid(x):
    return ( sigmoid(x)*(1-sigmoid(x)))
```

```python
x=np.linspace(-7, 7, 100)
f=sigmoid(x)
df=dsigmoid(x)
plt.plot(x,f, label="sigmoid")
plt.plot(x, df, label="derivate of sigmoid")
plt.legend()
plt.show()
```

## Loss Function : MSE 사용

### Update Parameters

$b_{new}=b_{old}-\eta\nabla_{b}L$

$w_{new}=w_{old}-\eta\nabla_{w}L$

$\nabla_{b}L=\delta$

$\nabla_{w}L=\delta \times x$

$\delta=(a-y)\sigma'(z)$

```python
z = 0.5*1 + 1.0
dsigmoid(z)
```

```python
def train(w0, b0, lr=0.5, epochs=300):
    w = w0
    b = b0
    outputs = np.array([])
    deltas = np.array([])
    for _ in range(epochs):
        z = w*x + b
        a = sigmoid(z)
        outputs = np.append(outputs, a)

        delta = (a-y)*dsigmoid(z)
        deltas = np.append(deltas, delta)

        w = w - lr*delta*x
        b = b - lr*delta
    plt.plot(np.arange(epochs), outputs)
    plt.ylabel("output")
    plt.xlabel("epoch")
    plt.show()
    
    return(w, b, outputs, deltas)
```
